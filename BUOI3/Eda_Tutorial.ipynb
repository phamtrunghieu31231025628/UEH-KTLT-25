```python
# ==============================
# Exploratory Data Analysis (EDA) in Python
# ==============================

# Introduction
# Exploratory Data Analysis of the famous Iris dataset.
# Features: Sepal Width, Sepal Length, Petal Width, Petal Length
# Target: "species" có 3 lớp: setosa, versicolor, virginica

# Initial EDA: dùng pandas, matplotlib, seaborn
# ==============================

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load dataset
df = sns.load_dataset("iris")

# Alternative cách khác (nếu cần sklearn):
"""
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
y = iris.target
iris_df = pd.DataFrame(X, columns=iris.feature_names)
iris_df['target'] = y
"""

# ==============================
# Non-graphical EDA
# ==============================

# Xem toàn bộ dataset
df

# Kích thước dataset
df.shape

# 5 dòng đầu tiên
df.head()

# Thông tin chi tiết
df.info()

# Kiểm tra missing values
df.isnull().sum()

# Thống kê cơ bản
df.describe()

# Kiểm tra balance của các lớp
df["species"].value_counts()

# Ma trận tương quan (chỉ lấy cột số)
df.corr(numeric_only=True)

# ==============================
# Graphical EDA
# ==============================

# Biểu đồ phân bố số lượng từng loài
sns.countplot(x="species", data=df)
plt.title("Class Distribution (Species)")
plt.show()

# Pairplot để xem mối quan hệ giữa các biến
sns.pairplot(df, hue="species")
plt.show()

# Heatmap của ma trận tương quan
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Ma trận tương quan (chỉ lấy cột số)
df.corr(numeric_only=True)
df.select_dtypes(include=["float64", "int64"]).corr()

# Dùng matplotlib để vẽ bar chart
import matplotlib.pyplot as plt

df["species"].value_counts().plot(kind="bar", color=["skyblue", "salmon", "lightgreen"])
plt.xlabel("Species")
plt.ylabel("Count")
plt.title("Number of samples per species")
plt.show()

# Bar chart số lượng từng loài
counts = df["species"].value_counts()
plt.bar(counts.index, counts.values, color=["skyblue", "salmon", "lightgreen"])
plt.xlabel("Species")
plt.ylabel("Count")
plt.title("Number of Samples per Species")
plt.show()

# Histogram ví dụ cho Sepal Length
plt.hist(df["sepal_length"], bins=10, color="skyblue", edgecolor="black")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Length")
plt.show()

# Sử dụng seaborn để vẽ countplot cho species
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="species", data=df, palette="Set2")
plt.title("Species Count (Seaborn)")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()

# Scatterplot dùng pandas
df.plot.scatter(x="sepal_length", y="sepal_width", alpha=0.7)
plt.title("Scatterplot of Sepal Length vs Sepal Width (Pandas)")
plt.show()

# Scatterplot dùng matplotlib
plt.scatter(df["sepal_length"], df["sepal_width"], alpha=0.7, c="skyblue", edgecolors="black")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Scatterplot of Sepal Length vs Sepal Width (Matplotlib)")
plt.show()

# Scatterplot dùng matplotlib với cú pháp truyền data trực tiếp
plt.scatter('sepal_length', 'sepal_width', data=df, alpha=0.7, c="orange", edgecolors="black")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Scatterplot of Sepal Length vs Sepal Width (Matplotlib - data arg)")
plt.show()

# Scatterplot cơ bản với matplotlib
plt.scatter(df['sepal_length'], df['sepal_width'], alpha=0.7, c="purple", edgecolors="black")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Scatterplot of Sepal Length vs Sepal Width")
plt.show()

# Scatterplot dùng plt.plot (tương đương scatter)
plt.plot('sepal_length', 'sepal_width', data=df, marker='.', linestyle='none')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Scatterplot of Sepal Length vs Sepal Width (plt.plot)")
plt.show()

# Scatterplot cơ bản với seaborn
sns.scatterplot(x="sepal_length", y="sepal_width", data=df)
plt.title("Scatterplot of Sepal Length vs Sepal Width (Seaborn)")
plt.show()

# Pairplot hiển thị scatterplot cho từng cặp biến
sns.pairplot(df, hue="species", diag_kind="kde", palette="Set1")
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()

# Pairplot có phân biệt species
sns.pairplot(df, hue="species", diag_kind="kde", palette="Set2")
plt.suptitle("Pairplot of Iris Dataset by Species", y=1.02)
plt.show()

# Tính toán ma trận tương quan
corr = df.select_dtypes(include=['float64', 'int64']).corr()

# Vẽ heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()

# Histogram với seaborn
plt.figure(figsize=(10,6))
sns.histplot(df['sepal_length'], kde=True, bins=20, color='skyblue')
plt.title("Distribution of Sepal Length")
plt.show()

# Boxplot
sns.boxplot(x="species", y="sepal_length", data=df)
plt.title("Boxplot of Sepal Length by Species")
plt.show()

# Violin plot
sns.violinplot(x="species", y="sepal_length", data=df)
plt.title("Violin Plot of Sepal Length by Species")
plt.show()

features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

plt.figure(figsize=(12,8))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)  # 2x2 grid
    sns.boxplot(x="species", y=feature, data=df)
    plt.title(f"Boxplot of {feature} by Species")

plt.tight_layout()
plt.show()

df_melt = df.melt(id_vars="species", var_name="feature", value_name="value")

plt.figure(figsize=(12,8))
sns.boxplot(x="feature", y="value", hue="species", data=df_melt)
plt.title("Boxplots of all features by Species")
plt.show()

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

sns.violinplot(x="species", y="sepal_length", data=df, ax=axes[0,0])
axes[0,0].set_title("Sepal Length")

sns.violinplot(x="species", y="sepal_width", data=df, ax=axes[0,1])
axes[0,1].set_title("Sepal Width")

sns.violinplot(x="species", y="petal_length", data=df, ax=axes[1,0])
axes[1,0].set_title("Petal Length")

sns.violinplot(x="species", y="petal_width", data=df, ax=axes[1,1])
axes[1,1].set_title("Petal Width")

plt.tight_layout()
plt.show()
```
